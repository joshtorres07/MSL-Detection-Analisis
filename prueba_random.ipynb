```python
import pandas as pd
import random
import numpy as np
# Cargar los datos del archivo CSV
df = pd.read_csv('https://raw.githubusercontent.com/joshtorres07/Big-Data-Project/main/Experimentacion_Victor.csv')

# Generar nuevos datos para las columnas solicitadas
letras = [chr(i) for i in range(97, 123)]  # Generar una lista de letras del alfabeto
df['SEÑA ESPERADA'] = np.random.choice(letras, size=len(df))
df['HUAWEI/5001'] = np.random.choice(letras, size=len(df))
df['DISTANCIA TOTAL'] = [random.uniform(0, 50) for _ in range(len(df))]
df['#INCIDENCIAS'] = [random.randint(0, 10) for _ in range(len(df))]
df['FALSAS'] = [random.randint(0, 5) for _ in range(len(df))]
df['SAMSUNG-B/5000'] = np.random.choice(letras, size=len(df))
df['DISTANCIA TOTAL.1'] = [random.uniform(0, 50) for _ in range(len(df))]
df['#INCIDENCIAS.1'] = [random.randint(0, 10) for _ in range(len(df))]
df['FALSAS.1'] = [random.randint(0, 5) for _ in range(len(df))]
df['S8/5003'] = np.random.choice(letras, size=len(df))
df['DISTANCIA TOTAL.2'] = [random.uniform(0, 50) for _ in range(len(df))]
df['#INCIDENCIAS.2'] = [random.randint(0, 10) for _ in range(len(df))]
df['FALSAS.2'] = [random.randint(0, 5) for _ in range(len(df))]

# Imprimir el DataFrame resultante
print(df)
```

```python
import pandas as pd
import numpy as np

# Cargar los datos del archivo CSV
df = pd.read_csv('Experimentacion_Rosario.csv')

# Agrupar los datos por valores únicos de las columnas numéricas
grouped = df.groupby(['HUAWEI/5001', 'DISTANCIA TOTAL', '#INCIDENCIAS', 'FALSAS', 'SAMSUNG-B/5000', 'DISTANCIA TOTAL.1', '#INCIDENCIAS.1', 'FALSAS.1', 'S8/5003', 'DISTANCIA TOTAL.2', '#INCIDENCIAS.2', 'FALSAS.2'])

# Crear un diccionario que asocie cada grupo con una letra aleatoria
letras = [chr(i) for i in range(65, 90)]  # Generar una lista de letras del alfabeto
group_keys = list(grouped.groups.keys())
random_letras = np.random.choice(letras, size=len(group_keys))
letras_dict = dict(zip(group_keys, random_letras))

# Asignar una letra aleatoria a cada fila del DataFrame
df['SEÑA ESPERADA'] = [letras_dict[tuple(row)] for row in df[['HUAWEI/5001', 'DISTANCIA TOTAL', '#INCIDENCIAS', 'FALSAS', 'SAMSUNG-B/5000', 'DISTANCIA TOTAL.1', '#INCIDENCIAS.1', 'FALSAS.1', 'S8/5003', 'DISTANCIA TOTAL.2', '#INCIDENCIAS.2', 'FALSAS.2']].values]

# Imprimir el DataFrame resultante
print(df)
df.to_csv("Experimentacion_Josue.csv", index=False)
```

```python
import pandas as pd
import numpy as np

# cargar el DataFrame original
df = pd.read_csv("Experimentacion_Victor.csv", encoding='latin-1')

# definir una función que agrega ruido a una columna numérica
def add_noise(col):
    if np.issubdtype(col.dtype, np.number):
        noise = np.random.rand(len(col)) * 22500 # ajusta el factor de ruido a tu gusto
        return col + noise
    else:
        return col

def add_noise_inc(col):
    if np.issubdtype(col.dtype, np.number):
        noise = np.random.rand(len(col)) * 0.6 # ajusta el factor de ruido a tu gusto
        return col + noise
    else:
        return col

# aplicar la función a las columnas de distancia e incidencia
df[["DISTANCIA TOTAL.1", "DISTANCIA TOTAL", 'DISTANCIA TOTAL.2']] = df[["DISTANCIA TOTAL.1", "DISTANCIA TOTAL", 'DISTANCIA TOTAL.2']].apply(add_noise)

df[['#INCIDENCIAS.1','#INCIDENCIAS','#INCIDENCIAS.2']] = df[['#INCIDENCIAS.1','#INCIDENCIAS','#INCIDENCIAS.2']].apply(add_noise_inc)
df[['#INCIDENCIAS.1','#INCIDENCIAS','#INCIDENCIAS.2']] = df[['#INCIDENCIAS.1','#INCIDENCIAS','#INCIDENCIAS.2']].round().astype(int)
# escribir los datos en un archivo CSV
df.to_csv("Experimentacion_Itzel.csv", index=False, encoding='latin-1')

```

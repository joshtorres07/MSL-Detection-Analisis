```python
import pandas as pd
import numpy as np

# cargar el DataFrame original
df = pd.read_csv("Experimentacion_Victor.csv", encoding='latin-1')

# definir una función que agrega ruido a una columna numérica
def add_noise(col):
    if np.issubdtype(col.dtype, np.number):
        noise = np.random.rand(len(col)) * 22500 # ajusta el factor de ruido a tu gusto
        return col + noise
    else:
        return col

def add_noise_inc(col):
    if np.issubdtype(col.dtype, np.number):
        noise = np.random.rand(len(col)) * 1.6  # ajusta el factor de ruido a tu gusto
        return col + noise
    else:
        return col

# aplicar la función a las columnas de distancia e incidencia
df[["DISTANCIA TOTAL.1", "DISTANCIA TOTAL", 'DISTANCIA TOTAL.2']] = df[["DISTANCIA TOTAL.1", "DISTANCIA TOTAL", 'DISTANCIA TOTAL.2']].apply(add_noise)

df[['#INCIDENCIAS.1','#INCIDENCIAS','#INCIDENCIAS.2']] = df[['#INCIDENCIAS.1','#INCIDENCIAS','#INCIDENCIAS.2']].apply(add_noise_inc)
df[['#INCIDENCIAS.1','#INCIDENCIAS','#INCIDENCIAS.2']] = df[['#INCIDENCIAS.1','#INCIDENCIAS','#INCIDENCIAS.2']].round().astype(int)
# escribir los datos en un archivo CSV
df.to_csv("Experimentacion_Daniel.csv", index=False, encoding='latin-1')
```

```python
import pandas as pd
import numpy as np

# cargar el DataFrame original
df = pd.read_csv("Experimentacion_Victor.csv", encoding='latin-1')

# iterar sobre cada celda en el DataFrame y agregar ruido
for i in range(len(df)):
    for j in range(len(df.columns)):
        # generar un valor aleatorio
        noise = np.random.rand() * 0.1 # ajusta el factor de ruido a tu gusto

        # agregar ruido a la celda
        if isinstance(df.iloc[i, j], (int, float)):
            df.iloc[i, j] += noise
        elif isinstance(df.iloc[i, j], str):
            df.iloc[i, j] += str(noise)
        else:
            pass # no agregar ruido a otros tipos de datos

# escribir los datos en un archivo CSV
df.to_csv("Experimentacion_Victor_ruidoso.csv", index=False)
```

```python
import pandas as pd
import numpy as np

# cargar el DataFrame original
df = pd.read_csv("Experimentacion_Victor.csv")

# definir una función que agrega ruido a una columna numérica
def add_noise(col):
    if np.issubdtype(col.dtype, np.number):
        noise = np.random.rand(len(col)) * 0.1 # ajusta el factor de ruido a tu gusto
        return col + noise
    else:
        return col

# aplicar la función a las columnas de distancia e incidencia
df[["distancia", "incidencia"]] = df[["distancia", "incidencia"]].apply(add_noise)

# escribir los datos en un archivo CSV
df.to_csv("Experimentacion_Victor_ruidoso.csv", index=False)
```
